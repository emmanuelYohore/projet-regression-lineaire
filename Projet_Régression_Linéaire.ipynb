### Importations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Données

```python
# LECTURE CSV à éxécuter

# lecture à partir de Github pour pouvoir éxécuter le notebook sans se soucier d'importer le fichier csv
url = 'https://raw.githubusercontent.com/moncoachdata/MasterClass_DS_ML/main/AMES_Final_DF.csv'
df = pd.read_csv(url)

# Sinon lecture classique en important sur Google Colab ou depuis le chemin exact
# df = pd.read_csv("AMES_Final_DF.csv")
```

### Création de X et y
**Tâche : Le label ou variable cible que nous essayons de prédire est la colonne SalePrice. Séparez les données en Features X et Label y**.

```python
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
```

### Fractionnement Entraînement Test
**Tâche : Utilisez scikit-learn pour séparer X et y en un ensemble d'entraînement et un ensemble de test. Comme nous utiliserons plus tard une stratégie de recherche par grille (grid search), fixez votre proportion de test à 10 %. Pour obtenir la même répartition des données que dans le notebook solutions, vous pouvez spécifier random_state = 101**.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,
                                                    random_state=101)
```

### Mise à l'échelle des données
**Tâche : Les features de l'ensemble de données ont une variété d'échelles et d'unités. Pour une performance de régression optimale, mettez à l'échelle les Features X. Prenez note de ce que vous utilisez pour .fit() et de ce que vous utilisez pour .transform()**.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

### Modèle
**Tâche : Nous allons utiliser un modèle Elastic Net. Créez une instance du modèle ElasticNet par défaut avec scikit-learn**.

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_model = ElasticNet()
```

**Tâche : Le modèle Elastic Net a deux paramètres principaux, alpha et le ratio L1 (l1_ratio). Créez un dictionnaire de grille de paramètres des valeurs pour l'Elastic Net. N'hésitez pas à jouer avec ces valeurs, gardez à l'esprit qu'elles peuvent ne pas correspondre exactement aux choix de la solution**.

```python
param_grid = {'alpha': [0.1, 1, 5, 10,100],
              'l1_ratio': [0.1,0.7, 0.99, 1]}
```

**Tâche : À l'aide de scikit-learn, créez un objet GridSearchCV et exécutez un Grid Search pour les meilleurs paramètres de votre modèle en fonction de vos données d'entraînement mises à l'échelle. [Au cas où vous seriez curieux de connaître les avertissements que vous pourriez recevoir pour certaines combinaisons de paramètres](https://stackoverflow.com/questions/20681864/lasso-on-sklearn-does-not-converge)**

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=1)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

**Tâche : Afficher la meilleure combinaison de paramètres pour votre modèle**.

```python
grid_model.best_params_
```

### Évaluation du modèle
**Tâche : Évaluez les performances de votre modèle sur les 10% de données non vues auparavant (ensemble de test mis à l'échelle)**.

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
np.mean(df['SalePrice'])
```

```python
R2 = r2_score(y_test,y_pred)
R2
```
